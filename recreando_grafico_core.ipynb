```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('tips')
df
df_unicos = df['time'].unique()
df_unicos

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper left', fontsize='large', shadow=True, title='Detalles')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')

# Añadir anotación
max_tip = df[df['time'] == 'Dinner']['tip'].max() #calcular el valor de tip maximo
max_tip_row = df[(df['time'] == 'Dinner') & (df['tip'] == max_tip)] #calcular la fila con el valor maximo de tip
total_bill_max_tip = max_tip_row['total_bill'].values[0] #extrae el valor del total bill en donde el tip fue maximo
tip_max_tip = max_tip_row['tip'].values[0] #extraer el valor del maximo tip

ax.annotate('Mayor tip en la cena', xy=(total_bill_max_tip, tip_max_tip), xytext=(total_bill_max_tip, tip_max_tip), #queremos resaltar donde está el mayor tip con la flecha
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha
ax.arrow(34, 8, 15, 2, head_width=0.5, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()

#Obs: Podemos observar que generalmente durante las cenas se dejan mayores propinas. El valor mas alto de propina fue durante una cena. Existe una relacion positiva entre total bill y tip, lo cua es lógico.
```
